```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('train_log.pickle', 'rb') as f:
    data = pickle.load(f, encoding='bytes')
```

```python
%matplotlib inline
x = range(len(data['a']['loss']) // 10 + 1)
plt.plot(x, [d for i, d in enumerate(data['a']['acc']) if i % 10 == 0], label='A accuracy')
plt.plot(x, [d for i, d in enumerate(data['d']['acc']) if i % 10 == 0], label='D accuracy')
plt.legend()
```
